```python
# miramos los nulos
def calculate_null_percentage(data):
    null_percentage = round(data.isna().sum() / len(data) * 100, 4)
    nulls_df = pd.DataFrame(null_percentage).reset_index()
    nulls_df.columns = ['header_name', 'percent_nulls']
    return nulls_df
```

```python
# Función para reemplazar los nans de Genero, Episode Duration y Rating
def reemplazar_nans(fila):
    # Si 'Genero' es NaN o vacío, cogemos el valor de 'Genre'
    if pd.isna(fila["Genero"]) or fila["Genero"] == 0:
        fila["Genero"] = fila["Genre"]
    
    # Si 'Episode Duration' es NaN o 0, cogemos el valor de 'Runtime'
    if pd.isna(fila["Episode Duration"]) or fila["Episode Duration"] == 0:
        fila["Episode Duration"] = fila["Runtime"]
    
    # Si 'Rating_x' es NaN, cogemos el valor de 'Rating_y'
    if (pd.isna(fila["Rating_x"]) or fila["Rating_x"] == 0) & pd.notna(fila["Rating_y"]):
        fila["Rating_x"] = fila["Rating_y"]
    
    return fila

# Aplicamos la función a cada fila del DataFrame
df_sinano = df_sinano.apply(reemplazar_nans, axis=1)
```
